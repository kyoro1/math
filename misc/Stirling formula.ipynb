```python
import math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## Stirling formula
$$n! \sim \sqrt{2\pi n}\times\big(\frac{n}{e}\big)^n, \rm{for \ }  n \gg 1$$

```python
def calc_prod(n):
    return math.sqrt(2 * math.pi * n) * ((n / math.e) ** n)

def calc_stirling_formula(n):
    return math.factorial(n) / calc_prod(n)
```

```python
list_stirling = [(n, calc_stirling_formula(n)) for n in range(1,150)]
```

```python
df_stirling = pd.DataFrame(list_stirling, columns = ['n', 'approximation'])
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.lineplot(data=df_stirling, x='n', y='approximation')
```
